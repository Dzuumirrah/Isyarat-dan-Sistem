## A.1 Karakterisitik Isyarat Berdasarkann Jenis Variabel

### A.1.1 Isyarat Waktu Kontinu

Isyarat waktu kontinyu adalah sebuah sinyal yang memiliki nilai kontinyu berdasarkan input waktu yang diberikan. Nilai tersebut umumnya adalah berupa interval tertentu seperti $ 0 > t > 5, t \leq 0,$ dan $ t = 0$.

Contohnya adalah <br>
$f(t) = 2\sin(t),$ untuk $2 < t < 5$<br>
$f(t) = 0,$ untuk $t$ lainnya.

```python
# Import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pembuatan grafik
# Shortcut
sin = np.sin
pi = np.pi

# deklarasi nilai setiap variabel
A = 2               # Amplitudo

# Setting plot untuk grafik
t = np.linspace(0, 10, 1000)
x_t = np.where((2 < t) & (t < 6), A * sin(t), 0) 

plt.figure(figsize=(15, 2))      # size dalam inci
plt.plot(t, x_t)
plt.title("Grafik fungsi $f(t) = 2\sin(t)$, dengan $ 2 < t < 6$ ")
plt.xlabel("waktu (s)")
plt.show()
```

### A.1.2 Isyarat Waktu diskrit

Isyarat waktu diskrit adalah sebuah sinyal yang memiliki nilai disikrit berdasarkan input waktu yang diberikan. Nilai tersebut umumnya adalah berupa pola tertentu seperti $ n = 1, 2, 3, ...$ atau $ n =$ bilangan prima.

Contohnya adalah <br>
$f[n] = 2\sin(n),$ untuk $n =$ bilangan prima kurang dari 100

```python
# Import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Algoritma bilangan prima
from math import sqrt

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

prime = []
for i in range(1, 100):
    if is_prime(i):
        prime.append(i)

# Pembuatan grafik
# Shortcut
sin = np.sin
pi = np.pi

# deklarasi nilai setiap variabel
A = 2               # Amplitudo

# Setting plot untuk grafik
n = np.arange(0, 20, 1)
x_n = np.array([A * sin(i) if i in prime else 0 for i in n])

plt.figure(figsize=(15, 2))      # size dalam inci
plt.stem(n, x_n)
plt.title("Grafik fungsi $f[n] = 2\sin(n)$, dengan $n =$ bilangan prima kurang dari 20 ")
plt.xlabel("waktu (s)")
plt.show()
```

### A.1.3 Isyarat Waktu Digital

Isyarat waktu digital adalah sebuah sinyal yang hanya memiliki dua kemungkinan nilai seperti $0 dan 1, 0 dan 64$ dan lain lain yang berkaitan dengan sistem bilangan digital. 

Contohnya adalah <br>
$x[n] = \cos(n)$ dalam interval $-a \leq n \leq a$ pada sinyal biner 

```python
# Pembuatan grafik
# shortcut
cos = np.cos
# Deklarasi nilai tiap variabel
a = 17
t_digit = np.arange(-a, a, 1)

# Setting plot untuk grafik
x_t_disc = cos(t_digit)
x_t_digit = np.where(x_t_disc >= 0, 1, 0)

# Display grafik
plt.figure(figsize=(10, 5))
plt.ylim(-2, 2)
plt.xlim(-18, 18)
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.stem(t_digit, x_t_digit, linefmt="blue", basefmt="none", markerfmt="blue")
plt.grid(True)
```

## A.2 Dasar Isyarat Waktu kontinu

### A.2.1 Isyarat Sinusoidal

**Isyarat waktu kontinyu sinusoidal** adalah sebuah sinyal yang memiliki bentuk grasik sinusoidal. Sinyal ini dapat dikontruksi dengan memasukkan fungsi sinusoidal pada sistem.

Contohnya adalah <br>
$f(t) = 2\sin(t),$ untuk $2 < t < 5$<br>
$f(t) = 0,$ untuk $t$ lainnya.

```python
# Pembuatan grafik
# Shortcut
sin = np.sin
pi = np.pi

# deklarasi nilai setiap variabel
A = 2               # Amplitudo

# Setting plot untuk grafik
t = np.linspace(0, 10, 1000)
x_t = np.where((2 < t) & (t < 6), A * sin(t), 0) 

plt.figure(figsize=(15, 2))      # size dalam inci
plt.plot(t, x_t)
plt.title("Grafik fungsi $f(t) = 2\sin(t)$, dengan $ 2 < t < 6$ ")
plt.xlabel("waktu (s)")
plt.show()
```

### A.2.2 Isyarat Eksponensial Kompleks

**Isyarat Waktu Kontinu Eksponensial Kompleks** adalah sebuah sinyal waktu dengan bentuk fungsi eksponensial yang memiliki bilangan kompleks di dalamnya. Secara umum, sinyal waktu kontinu eksponensial memiliki bentuk sebagai berikut:

$f(t) = C e^{at}$

Pada bentuk kompleksnya, $a$ merupakan bilangan imajiner dengan $a=j\omega _0$<br>
Contohnya adalah <br>
$f(t) = 3e^{j\omega t},$ dengan $ 0 < t < 5$ dan $f=1 Hz$

```python
# Pembuatan grafik
# shortcut
exp = np.exp

# variabel
A = 3                   # Amplitudo
f = 1                   # frekuensi (Hz)
omega = 2 * pi * f      # kecepatan sudut
t = np.linspace(0, 5, 1000) 

# Plot fungsi
x_t = A * exp(1j * omega * t)

# Plot 
plt.figure(figsize=(10, 6))

# Real part
plt.subplot(2, 1, 1)
plt.plot(t, np.real(x_t), label="Bagian real", color='b')
plt.title("Isyarat Waktu Kontinu Eksponensial Kompleks")
plt.ylabel("Bagian real")
plt.grid(True)
plt.legend()

# Imaginary part
plt.subplot(2, 1, 2)
plt.plot(t, np.imag(x_t), label="Bagian imajiner", color='r')
plt.ylabel("Bagian imejiner")
plt.xlabel("waktu (s)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

Terlihat bahwa sinyal ini mirip dengan sinyal sinusoidal. Hal ini dapat terjadi karena bilangan eksponensial kompleks dapat berupa gabungan dari sinyal $\cos$ dan $\sin$

$f(t) = C e^{j\omega t} = C(\cos \omega t + j \sin \omega t)$

### A.2.3 Isyarat Waktu Kontinu Eksponensial

**Isyarat Waktu Kontinu Eksponensial** adalah sebuah sinyal waktu dengan bentuk fungsi eksponensial. Secara umum, sinyal waktu kontinu eksponensial memiliki bentuk sebagai berikut:

$f(t) = C e^{at}$

Walaupun terlihat mirip, sinyal waktu kontinu eksponensial memiliki bentuk grafik yang berbeda dengan fungsi eksponensial kompleks. 

Saat $a$ bernilai positif, tren grafik akan terlihat menaik. Namun, saat $a$ bernilai negatif, tren grafik akan terlihat menurun. 

Contohnya adalah <br>
$f(t) = 10e^{-2t},$ dengan $ 0 < t < 5$ dan<br>
$f(t) = 10e^{2t},$ dengan $ 0 < t < 5$

```python
# Pembuatan grafik
# Deklarasi nilai tiap variabel
C = 10
t = np.linspace(0, 5, 1000)

# Setting plot untuk grafik
x_t1 = C * exp(-2 * t)
x_t2 = C * exp(2 * t)

# Display grafik
plt.figure(figsize=(10, 15))

# grafik a negatif
plt.subplot(3, 1, 1)
plt.title("Grafik Sinyal Waktu Kontinu Eksponensial")
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.plot(t, x_t1, color="r", label= "a bernilai negatif")
plt.legend()
plt.grid(True)

# grafik a positif
plt.subplot(3, 1, 2)
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.plot(t, x_t2, color="b", label= "a bernilai positif")
plt.legend()
plt.grid(True)

# grafik gabungan
plt.subplot(3, 1, 3)
plt.title("Gabungan kedua grafik")
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.ylim(0, 100)
plt.plot(t, x_t2, color="b", label= "a bernilai positif")
plt.plot(t, x_t1, color="r", label= "a bernilai negatif")
plt.legend()
plt.grid(True)


plt.tight_layout()
plt.show()

```

### A.2.4 Isyarat Waktu Kontinu Unit Step

**Isyarat Waktu Kontinu Unit Step** adalah sebuah sinyal waktu yang akan memiliki nilai (umumnya bernilai 1) saat $t \geq 0$. Secara umum, sinyal waktu kontinu unist step memiliki bentuk sebagai berikut:

$U(t) = 0; t < 0$<br>
$U(t) = 1; t \geq 0$

```python
# Pembuatan grafik
# Deklarasi nilai tiap variabel
t = np.linspace(-3, 5, 1000)

# Setting plot untuk grafik
x_t_step = np.where(t >= 0, 1, 0)
# Display

plt.figure(figsize=(10, 5))
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.title("Sinyal Waktu Kontinu Unit Step saat $t \geq 0$")
plt.ylim(-1, 3)
plt.plot(t, x_t_step, color="blue", label="$U(t)_{step}$")

plt.grid(True)
plt.legend()

plt.show()
```

### A.2.5 Isyarat Waktu Kontinu Unit Impuls atau *Dirac Delta*

**Isyarat Waktu Kontinu Unit Impuls** adalah sebuah sinyal waktu yang akan memiliki nilai (umumnya bernilai 1) pada saat $t$ tertentu. Munculnya nilai ini bersifat sesaat ($t$ mendekati 0). 

***Dirac Delta*** bukanlah sebuah fungsi reguler, melainkan sebuah distribusi atau "fungsi yang digeneralisasi" yang digunakan untuk membuat sebuah distribusi nilai pada sebuah grafik dengan nilai 0 pada seluruh grafik kecuali pada titik 0 grafik. Secara umum, sinyal waktu kontinu unit impuls memiliki bentuk sebagai berikut:

$(t) = 0; t$ lainnya<br>
$(t) = A; t = 0 \geq 0$

Atau jika menggunakan persamaan *Dirac Delta*<br>
$(t) = \frac{1}{\lvert a \rvert \sqrt( \pi)} e^{-(\frac{t}{a})^2}$

dengan $A=$ amplitudo impuls.

```python
# Pembuatan grafik
# shortcut
sqrt = np.sqrt
pi = np.pi
exp = np.exp

# Deklarasi nilai tiap variabel
t = np.linspace(-5, 5, 1000)
a = 0.1    # semakin kecil, semakin runcing impuls

# Setting plot untuk grafik
_t = exp(-(t/a)**2) / (a * sqrt(pi))
# Display

plt.figure(figsize=(10, 5))
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.title("Pendekatan Sinyal Waktu Kontinu Unit Impuls Dirac Delta dengan Fungsi Gaussian")
plt.plot(t, _t, color="blue", label="$_t$")

plt.grid(True)
plt.legend()

plt.show()
```
