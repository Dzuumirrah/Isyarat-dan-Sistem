# A. Memahami Dasar Isyarat

## A.1 Karakterisitik Isyarat Berdasarkann Jenis Variabel

### A.1.1 Isyarat Waktu Kontinu

Isyarat waktu kontinyu adalah sebuah sinyal yang memiliki nilai kontinyu berdasarkan input waktu yang diberikan. Nilai tersebut umumnya adalah berupa interval tertentu seperti $ 0 > t > 5, t \leq 0,$ dan $ t = 0$.

Contohnya adalah <br>
$f(t) = 2\sin(t),$ untuk $2 < t < 5$<br>
$f(t) = 0,$ untuk $t$ lainnya.

```python
# Import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pembuatan grafik
# Shortcut
sin = np.sin
pi = np.pi

# deklarasi nilai setiap variabel
A = 2               # Amplitudo

# Setting plot untuk grafik
t = np.linspace(0, 10, 1000)
x_t = np.where((2 < t) & (t < 6), A * sin(t), 0) 

plt.figure(figsize=(15, 2))      # size dalam inci
plt.plot(t, x_t)
plt.title("Grafik fungsi $f(t) = 2\sin(t)$, dengan $ 2 < t < 6$ ")
plt.xlabel("waktu (s)")
plt.show()
```

### A.1.2 Isyarat Waktu diskrit

Isyarat waktu diskrit adalah sebuah sinyal yang memiliki nilai disikrit berdasarkan input waktu yang diberikan. Nilai tersebut umumnya adalah berupa pola tertentu seperti $ n = 1, 2, 3, ...$ atau $ n =$ bilangan prima.

Contohnya adalah <br>
$f[n] = 2\sin(n),$ untuk $n =$ bilangan prima kurang dari 100

```python
# Import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Algoritma bilangan prima
from math import sqrt

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

prime = []
for i in range(1, 100):
    if is_prime(i):
        prime.append(i)

# Pembuatan grafik
# Shortcut
sin = np.sin
pi = np.pi

# deklarasi nilai setiap variabel
A = 2               # Amplitudo

# Setting plot untuk grafik
n = np.arange(0, 20, 1)
x_n = np.array([A * sin(i) if i in prime else 0 for i in n])

plt.figure(figsize=(15, 2))      # size dalam inci
plt.stem(n, x_n)
plt.title("Grafik fungsi $f[n] = 2\sin(n)$, dengan $n =$ bilangan prima kurang dari 20 ")
plt.xlabel("waktu (s)")
plt.show()
```

## A.2 Dasar Isyarat Waktu kontinu

### A.2.1 Isyarat Sinusoidal

**Isyarat waktu kontinyu sinusoidal** adalah sebuah sinyal yang memiliki bentuk grafik sinusoidal. Sinyal ini dapat dikontruksi dengan memasukkan fungsi sinusoidal pada sistem.

Contohnya adalah <br>
$f(t) = 2\sin(t),$ untuk $2 < t < 5$<br>
$f(t) = 0,$ untuk $t$ lainnya.

```python
# Pembuatan grafik
# Shortcut
sin = np.sin
pi = np.pi

# deklarasi nilai setiap variabel
A = 2               # Amplitudo

# Setting plot untuk grafik
t = np.linspace(0, 10, 1000)
x_t = np.where((2 < t) & (t < 6), A * sin(t), 0) 

plt.figure(figsize=(15, 2))      # size dalam inci
plt.plot(t, x_t)
plt.title("Grafik fungsi $f(t) = 2\sin(t)$, dengan $ 2 < t < 6$ ")
plt.xlabel("waktu (s)")
plt.show()
```

### A.2.2 Isyarat Eksponensial Kompleks

**Isyarat Waktu Kontinu Eksponensial Kompleks** adalah sebuah sinyal waktu dengan bentuk fungsi eksponensial yang memiliki bilangan kompleks di dalamnya. Secara umum, sinyal waktu kontinu eksponensial memiliki bentuk sebagai berikut:

$f(t) = C e^{at}$

Pada bentuk kompleksnya, $a$ merupakan bilangan imajiner dengan $a=j\omega _0$<br>
Contohnya adalah <br>
$f(t) = 3e^{j\omega t},$ dengan $ 0 < t < 5$ dan $f=1 Hz$

```python
# Pembuatan grafik
# shortcut
exp = np.exp

# variabel
A = 3                   # Amplitudo
f = 1                   # frekuensi (Hz)
omega = 2 * pi * f      # kecepatan sudut
t = np.linspace(0, 5, 1000) 

# Plot fungsi
x_t = A * exp(1j * omega * t)

# Plot 
plt.figure(figsize=(10, 6))

# Real part
plt.subplot(2, 1, 1)
plt.plot(t, np.real(x_t), label="Bagian real", color='b')
plt.title("Isyarat Waktu Kontinu Eksponensial Kompleks")
plt.ylabel("Bagian real")
plt.grid(True)
plt.legend()

# Imaginary part
plt.subplot(2, 1, 2)
plt.plot(t, np.imag(x_t), label="Bagian imajiner", color='r')
plt.ylabel("Bagian imejiner")
plt.xlabel("waktu (s)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

Terlihat bahwa sinyal ini mirip dengan sinyal sinusoidal. Hal ini dapat terjadi karena bilangan eksponensial kompleks dapat berupa gabungan dari sinyal $\cos$ dan $\sin$

$f(t) = C e^{j\omega t} = C(\cos \omega t + j \sin \omega t)$

### A.2.3 Isyarat Waktu Kontinu Eksponensial

**Isyarat Waktu Kontinu Eksponensial** adalah sebuah sinyal waktu dengan bentuk fungsi eksponensial. Secara umum, sinyal waktu kontinu eksponensial memiliki bentuk sebagai berikut:

$f(t) = C e^{at}$

Walaupun terlihat mirip, sinyal waktu kontinu eksponensial memiliki bentuk grafik yang berbeda dengan fungsi eksponensial kompleks. 

Saat $a$ bernilai positif, tren grafik akan terlihat menaik. Namun, saat $a$ bernilai negatif, tren grafik akan terlihat menurun. 

Contohnya adalah <br>
$f(t) = 10e^{-2t},$ dengan $ 0 < t < 5$ dan<br>
$f(t) = 10e^{2t},$ dengan $ 0 < t < 5$

```python
# Pembuatan grafik
# Deklarasi nilai tiap variabel
C = 10
t = np.linspace(0, 5, 1000)

# Setting plot untuk grafik
x_t1 = C * exp(-2 * t)
x_t2 = C * exp(2 * t)

# Display grafik
plt.figure(figsize=(10, 15))

# grafik a negatif
plt.subplot(3, 1, 1)
plt.title("Grafik Sinyal Waktu Kontinu Eksponensial")
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.plot(t, x_t1, color="r", label= "a bernilai negatif")
plt.legend()
plt.grid(True)

# grafik a positif
plt.subplot(3, 1, 2)
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.plot(t, x_t2, color="b", label= "a bernilai positif")
plt.legend()
plt.grid(True)

# grafik gabungan
plt.subplot(3, 1, 3)
plt.title("Gabungan kedua grafik")
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.ylim(0, 100)
plt.plot(t, x_t2, color="b", label= "a bernilai positif")
plt.plot(t, x_t1, color="r", label= "a bernilai negatif")
plt.legend()
plt.grid(True)


plt.tight_layout()
plt.show()

```

### A.2.4 Isyarat Waktu Kontinu Unit Step

**Isyarat Waktu Kontinu Unit Step** adalah sebuah sinyal waktu yang akan memiliki nilai (umumnya bernilai 1) saat $t \geq 0$. Secara umum, sinyal waktu kontinu unist step memiliki bentuk sebagai berikut:

$U(t) = 0; t < 0$<br>
$U(t) = 1; t \geq 0$

```python
# Pembuatan grafik
# Deklarasi nilai tiap variabel
t = np.linspace(-3, 5, 1000)

# Setting plot untuk grafik
x_t_step = np.where(t >= 0, 1, 0)
# Display

plt.figure(figsize=(10, 5))
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.title(r"Sinyal Waktu Kontinu Unit Step saat $t \geq 0$")
plt.ylim(-1, 3)
plt.plot(t, x_t_step, color="blue", label="$U(t)_{step}$")

plt.grid(True)
plt.legend()

plt.show()
```

### A.2.5 Isyarat Waktu Kontinu Unit Impuls atau *Dirac Delta*

**Isyarat Waktu Kontinu Unit Impuls** adalah sebuah sinyal waktu yang akan memiliki nilai (seperti bernilai 1) pada saat $t$ tertentu. Munculnya nilai ini bersifat sesaat ($t$ mendekati 0). 

***Dirac Delta*** bukanlah sebuah fungsi reguler, melainkan sebuah distribusi atau "fungsi yang digeneralisasi" yang digunakan untuk membuat sebuah distribusi nilai pada sebuah grafik dengan nilai 0 pada seluruh grafik kecuali pada titik 0 grafik. Secara umum, sinyal waktu kontinu unit impuls memiliki bentuk seperti berikut ini:

$(t) = 0; t$ lainnya<br>
$(t) = A; t = 0$

Atau jika menggunakan persamaan *Dirac Delta*<br>
$(t) = \frac{1}{\lvert a \rvert \sqrt( \pi)} e^{-(\frac{t}{a})^2}$

dengan $A=$ amplitudo impuls.

```python
# Pembuatan grafik
# shortcut
sqrt = np.sqrt
pi = np.pi
exp = np.exp

# Deklarasi nilai tiap variabel
t = np.linspace(-5, 5, 1000)
a = 0.1    # semakin kecil, semakin runcing impuls

# Setting plot untuk grafik
_t = exp(-(t/a)**2) / (a * sqrt(pi))
# Display

plt.figure(figsize=(10, 5))
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.title("Pendekatan Sinyal Waktu Kontinu Unit Impuls Dirac Delta dengan Fungsi Gaussian")
plt.plot(t, _t, color="blue", label="$_t$")

plt.grid(True)
plt.legend()

plt.show()
```

# B. Properti Isyarat dan Transformasinya

## B.1 Properti dan Isyarat Waktu Kontinyu

### B.1.1 Isyarat Periodik Waktu Kontinyu

**Isyarat periodik waktu kontinyu** adalah sebuah sinyal yang memiliki nilai kontinyu berdasarkan input waktu yang diberikan dan memiliki bentuk periodik. Sinyal ini akan memiliki pola pada waktu periodik tertentu. Nilai tersebut umumnya adalah berupa interval tanpa terputus.

Contohnya adalah <br>
$f(t) = 2\sin(t)$

```python
# Pembuatan grafik
# List warna untuk tampilan garis
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

# deklarasi nilai setiap variabel
A = 2               # Amplitudo
T = 2 * pi          # Periode

# Setting plot untuk grafik
t = np.linspace(0, 30, 1000)
x_t = A * sin(t) 

plt.figure(figsize=(15, 5))      # size dalam inci
plt.title(r"Grafik fungsi $f(t) = 2\sin(t)$, dengan $ 2 < t < 6$ ")
plt.ylim(-3, 3)
for i in range (1, 5, 1):
    plt.axvline(x=T * i, color=colors[i-1], linestyle="--", label=f" batas periode ke-{i}")
plt.xlabel("waktu (s)")
plt.plot(t, x_t)

plt.legend()
plt.grid(True)
plt.show()
```

Terlihat bahwa grafik tersebut memiliki pola saat periode tertentu yang dipisahkan oleh garis putus-putus.

### B.1.2 Penjumlahan dari Isyarat Waktu Periodik Waktu Kontinyu

Sebuah penjumlahan isyarat waktu kontinyu akan menghasilkan grafik baru yang berbeda dengan isyarat sebelumnya. Hasil dari penjumlahan tersebut akan menyesuaikan dengan penjumlah yang digunakan. Umumnya, hasil penjumlahan isyarat waktu hanya mengubah amplitudo dari grafik saja, tidak dengan frekuensi dan periodenya. 

Contohnya adalah pejumlahan dari $f_1(t)$ dan $f_2(t)$ berikut yang mengashilkan $h(t)$<br>
$f_1(t) = \cos(t)$<br>
$f_2(t) = \sin(3t)$<br>

```python
# Pembuatan grafik
# shortcut
cos = np.cos

# Deklarasi variabel
T = 2 * pi
t = np.linspace(0, 20, 1000)

# Plot sinyal 
f_1t = cos(t)
f_2t = sin(3 * t)
h_t = cos(t) + sin(3 * t)

# Plotting
plt.figure(figsize=(15, 10))

# Grafik f_1(t)
plt.subplot(3, 1, 1)
plt.title(r'Sinyal $f_1(t) = \cos(t)$')
plt.plot(t, f_1t, label=r'$f_1(t) = cos(t)$', color="b")
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo')
plt.ylim(-3, 3)
plt.xlim(-.5, 20.5)
plt.grid(True)

# Grafik f_2(t)
plt.subplot(3, 1, 2)
plt.title(r'Sinyal $f_2(t) = \sin(3t)$')
plt.plot(t, f_2t, label=r'$f_2(t) = \sin(3t)$', color="g")
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo')
plt.ylim(-3, 3)
plt.xlim(-.5, 20.5)
plt.grid(True)

# Grafik h(t)
plt.subplot(3, 1, 3)
plt.title(r'Sinyal $h(t) = f_1(t) + f_2(t)$')
plt.plot(t, h_t, label=r'$h(t) = f_1(t) + f_2(t)$', color="r")
plt.axvline(x=T, color='r', linestyle='--', label='Period ke-1')
plt.axvline(x=2*T, color='g', linestyle='--', label='Period ke-2')
plt.axvline(x=3*T, color='b', linestyle='--', label='Period ke-3')
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo')
plt.ylim(-3, 3)
plt.xlim(-.5, 20.5)
plt.grid(True)

plt.tight_layout()
plt.legend()
plt.show()

```

### B.1.4  Waktu Kontinu dan Isyarat Casual (causal?) Waktu Diskrit 

Penulis bingung dengan soal karena keambiguan kalimat. Namun, akan diusahakan untuk menginterpretasi soal di atas sebaik mungkin.

**Isyarat kausal waktu diskrit dan kontinyu** Sistem disebut kausal jika output pada waktu $( t )$ hanya bergantung pada input di waktu sekarang atau waktu sebelumnya, yaitu $( t \geq 0 )$. Artinya, sistem tidak memperhitungkan nilai input di masa depan $(t\leq 0)$. Misalnya, jika $( x(t) )$ didefinisikan untuk $( t \geq 0 )$, output sistem bergantung pada nilai $( x )$ dari waktu sekarang dan sebelumnya.

Contohnya adalah (kausal diskrit) <br>
$f[n] = 2\sin[n], n = 0,1,2,3, ...$ 

```python
# Pembuatan grafik

# deklarasi nilai setiap variabel
A = 2               # Amplitudo
T = 2 * pi          # Periode

# Setting plot untuk grafik
n = np.arange(0, 30, 1)
x_n = A * sin(n) 

plt.figure(figsize=(15, 5))      # size dalam inci
plt.title(r"Grafik fungsi $f[n] = 2\sin(n)$, dengan $n=1,2,3,...$ ")
plt.ylim(-3, 3)
plt.xlabel("waktu (s)")
plt.stem(n, x_n, label=r"$f [n]=2\sin(n)$")

plt.legend()
plt.grid(True)
plt.show()
```

### B.1.5 Penjumlahan dari Isyarat Periodik Waktu Diskrit Berkala

Sebuah penjumlahan isyarat waktu diskrit akan menghasilkan grafik baru yang berbeda dengan isyarat sebelumnya. Hasil dari penjumlahan tersebut akan menyesuaikan dengan penjumlah yang digunakan. Umumnya, hasil penjumlahan isyarat waktu hanya mengubah amplitudo dari grafik saja, tidak dengan frekuensi dan periodenya. 

Contohnya adalah pejumlahan dari $f_1[n]$ dan $f_[n]$ berikut yang mengashilkan $h[n]$<br>
$f_1[n] = \cos(n)$<br>
$f_2[n] = \sin(3n)$<br>

```python
# Pembuatan grafik
# shortcut
cos = np.cos

# Deklarasi variabel
T = 2 * pi
t = np.arange(0, 20, 1)

# Plot sinyal 
f_1t = cos(t)
f_2t = sin(3 * t)
h_t = cos(t) + sin(3 * t)

# Plotting
plt.figure(figsize=(15, 10))

# Grafik f_1(t)
plt.subplot(3, 1, 1)
plt.title(r'Sinyal $f_1[n] = \cos(n)$')
plt.stem(t, f_1t, label=r'$f_1(t) = cos(t)$', linefmt="b", basefmt="#000000")
plt.xlabel('Urutan (n)')
plt.ylabel('Amplitudo')
plt.ylim(-3, 3)
plt.xlim(-.5, 20.5)
plt.grid(True)

# Grafik f_2(t)
plt.subplot(3, 1, 2)
plt.title(r'Sinyal $f_2[n] = \sin(3n)$')
plt.stem(t, f_2t, label=r'$f_2(t) = \sin(3t)$', linefmt="g", basefmt="#000000")
plt.xlabel('Urutan (n)')
plt.ylabel('Amplitudo')
plt.ylim(-3, 3)
plt.xlim(-.5, 20.5)
plt.grid(True)

# Grafik h(t)
plt.subplot(3, 1, 3)
plt.title(r'Sinyal $h[n] = f_1([n] + f_2[n]$')
plt.stem(t, h_t, label=r'$h(t) = f_1(t) + f_2(t)$', linefmt="r", basefmt="#000000")
plt.axvline(x=T, color='r', linestyle='--', label='Period ke-1')
plt.axvline(x=2*T, color='g', linestyle='--', label='Period ke-2')
plt.axvline(x=3*T, color='b', linestyle='--', label='Period ke-3')
plt.xlabel('Urutan (n)')
plt.ylabel('Amplitudo')
plt.ylim(-3, 3)
plt.xlim(-.5, 20.5)
plt.grid(True)

plt.tight_layout()
plt.legend()
plt.show()

```

### B.2.6 Isyarat Ganjil dan Genap

Agar sinyal dikatakan ganjil atau genap, salah satu dari kriteria di bawah harus terpenuhi:<br>
Sinyal genap: $x(t) = x(-t)$; Umumnya simetri terhadap sumbu y tertentu.<br>
Sinyal ganjil: $x(t) = -x(-t)$; Umumnya simetri terbalik terhadap sumbu y.<br>

```python
# Pembuatan grafik
# deklarasi variabel
t = np.linspace(-20, 20, 1000)

# Sinyal x(t) = t^2
x_t2 = t**2

# Sinyal x(t) = t^3
x_t3 = t**3

# Plotting
plt.figure(figsize=(10, 6))

# Plot Even signal
plt.subplot(2, 1, 1)
plt.plot(t, x_t2, label=r'$x(t) = t^2$', color="red")
plt.title(r'Sinyal Genap $x(t) = t^2$')
plt.ylim(-100, 100)
plt.axvline(0, color="#000000", linestyle="--")
plt.grid(True)
plt.legend()

# Plot Odd signal
plt.subplot(2, 1, 2)
plt.plot(t, x_t3, label=r'$y(t) = t^3$', color="blue")
plt.title(r'Sinyal Ganjil $y(t) = t^3$')
plt.ylim(-100, 100)
plt.axvline(0, color="#000000", linestyle="--")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

Terlihat bahwa fungsi $x(t)$ simetri terhadap sumbu y sehingga fungsi tersebut adalah sinyal genap. 

Lalu, fungsi $y(t)$ simetri dengan kondisi terbalik terhadap sumbu y sehingga fungsi tersebut adalah sinyal genap.

## A.2 Dasar Isyarat Waktu kontinu

### A.2.1 Isyarat Sinusoidal

**Isyarat waktu kontinyu sinusoidal** adalah sebuah sinyal yang memiliki bentuk grafik sinusoidal. Sinyal ini dapat dikontruksi dengan memasukkan fungsi sinusoidal pada sistem.

Contohnya adalah <br>
$f(t) = 2\sin(t),$ untuk $2 < t < 5$<br>
$f(t) = 0,$ untuk $t$ lainnya.

```python
# Pembuatan grafik
# Shortcut
sin = np.sin
pi = np.pi

# deklarasi nilai setiap variabel
A = 2               # Amplitudo

# Setting plot untuk grafik
t = np.linspace(0, 10, 1000)
x_t = np.where((2 < t) & (t < 6), A * sin(t), 0) 

plt.figure(figsize=(15, 2))      # size dalam inci
plt.plot(t, x_t)
plt.title("Grafik fungsi $f(t) = 2\sin(t)$, dengan $ 2 < t < 6$ ")
plt.xlabel("waktu (s)")
plt.show()
```

### A.2.2 Isyarat Eksponensial Kompleks

**Isyarat Waktu Kontinu Eksponensial Kompleks** adalah sebuah sinyal waktu dengan bentuk fungsi eksponensial yang memiliki bilangan kompleks di dalamnya. Secara umum, sinyal waktu kontinu eksponensial memiliki bentuk sebagai berikut:

$f(t) = C e^{at}$

Pada bentuk kompleksnya, $a$ merupakan bilangan imajiner dengan $a=j\omega _0$<br>
Contohnya adalah <br>
$f(t) = 3e^{j\omega t},$ dengan $ 0 < t < 5$ dan $f=1 Hz$

```python
# Pembuatan grafik
# shortcut
exp = np.exp

# variabel
A = 3                   # Amplitudo
f = 1                   # frekuensi (Hz)
omega = 2 * pi * f      # kecepatan sudut
t = np.linspace(0, 5, 1000) 

# Plot fungsi
x_t = A * exp(1j * omega * t)

# Plot 
plt.figure(figsize=(10, 6))

# Real part
plt.subplot(2, 1, 1)
plt.plot(t, np.real(x_t), label="Bagian real", color='b')
plt.title("Isyarat Waktu Kontinu Eksponensial Kompleks")
plt.ylabel("Bagian real")
plt.grid(True)
plt.legend()

# Imaginary part
plt.subplot(2, 1, 2)
plt.plot(t, np.imag(x_t), label="Bagian imajiner", color='r')
plt.ylabel("Bagian imejiner")
plt.xlabel("waktu (s)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

Terlihat bahwa sinyal ini mirip dengan sinyal sinusoidal. Hal ini dapat terjadi karena bilangan eksponensial kompleks dapat berupa gabungan dari sinyal $\cos$ dan $\sin$

$f(t) = C e^{j\omega t} = C(\cos \omega t + j \sin \omega t)$

### A.2.3 Isyarat Waktu Kontinu Eksponensial

**Isyarat Waktu Kontinu Eksponensial** adalah sebuah sinyal waktu dengan bentuk fungsi eksponensial. Secara umum, sinyal waktu kontinu eksponensial memiliki bentuk sebagai berikut:

$f(t) = C e^{at}$

Walaupun terlihat mirip, sinyal waktu kontinu eksponensial memiliki bentuk grafik yang berbeda dengan fungsi eksponensial kompleks. 

Saat $a$ bernilai positif, tren grafik akan terlihat menaik. Namun, saat $a$ bernilai negatif, tren grafik akan terlihat menurun. 

Contohnya adalah <br>
$f(t) = 10e^{-2t},$ dengan $ 0 < t < 5$ dan<br>
$f(t) = 10e^{2t},$ dengan $ 0 < t < 5$

```python
# Pembuatan grafik
# Deklarasi nilai tiap variabel
C = 10
t = np.linspace(0, 5, 1000)

# Setting plot untuk grafik
x_t1 = C * exp(-2 * t)
x_t2 = C * exp(2 * t)

# Display grafik
plt.figure(figsize=(10, 15))

# grafik a negatif
plt.subplot(3, 1, 1)
plt.title("Grafik Sinyal Waktu Kontinu Eksponensial")
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.plot(t, x_t1, color="r", label= "a bernilai negatif")
plt.legend()
plt.grid(True)

# grafik a positif
plt.subplot(3, 1, 2)
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.plot(t, x_t2, color="b", label= "a bernilai positif")
plt.legend()
plt.grid(True)

# grafik gabungan
plt.subplot(3, 1, 3)
plt.title("Gabungan kedua grafik")
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.ylim(0, 100)
plt.plot(t, x_t2, color="b", label= "a bernilai positif")
plt.plot(t, x_t1, color="r", label= "a bernilai negatif")
plt.legend()
plt.grid(True)


plt.tight_layout()
plt.show()

```

### A.2.4 Isyarat Waktu Kontinu Unit Step

**Isyarat Waktu Kontinu Unit Step** adalah sebuah sinyal waktu yang akan memiliki nilai (umumnya bernilai 1) saat $t \geq 0$. Secara umum, sinyal waktu kontinu unist step memiliki bentuk sebagai berikut:

$U(t) = 0; t < 0$<br>
$U(t) = 1; t \geq 0$

```python
# Pembuatan grafik
# Deklarasi nilai tiap variabel
t = np.linspace(-3, 5, 1000)

# Setting plot untuk grafik
x_t_step = np.where(t >= 0, 1, 0)
# Display

plt.figure(figsize=(10, 5))
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.title("Sinyal Waktu Kontinu Unit Step saat $t \geq 0$")
plt.ylim(-1, 3)
plt.plot(t, x_t_step, color="blue", label="$U(t)_{step}$")

plt.grid(True)
plt.legend()

plt.show()
```

### A.2.5 Isyarat Waktu Kontinu Unit Impuls atau *Dirac Delta*

**Isyarat Waktu Kontinu Unit Impuls** adalah sebuah sinyal waktu yang akan memiliki nilai (seperti bernilai 1) pada saat $t$ tertentu. Munculnya nilai ini bersifat sesaat ($t$ mendekati 0). 

***Dirac Delta*** bukanlah sebuah fungsi reguler, melainkan sebuah distribusi atau "fungsi yang digeneralisasi" yang digunakan untuk membuat sebuah distribusi nilai pada sebuah grafik dengan nilai 0 pada seluruh grafik kecuali pada titik 0 grafik. Secara umum, sinyal waktu kontinu unit impuls memiliki bentuk seperti berikut ini:

$(t) = 0; t$ lainnya<br>
$(t) = A; t = 0$

Atau jika menggunakan persamaan *Dirac Delta*<br>
$(t) = \frac{1}{\lvert a \rvert \sqrt( \pi)} e^{-(\frac{t}{a})^2}$

dengan $A=$ amplitudo impuls.

```python
# Pembuatan grafik
# shortcut
sqrt = np.sqrt
pi = np.pi
exp = np.exp

# Deklarasi nilai tiap variabel
t = np.linspace(-5, 5, 1000)
a = 0.1    # semakin kecil, semakin runcing impuls

# Setting plot untuk grafik
_t = exp(-(t/a)**2) / (a * sqrt(pi))
# Display

plt.figure(figsize=(10, 5))
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.title("Pendekatan Sinyal Waktu Kontinu Unit Impuls Dirac Delta dengan Fungsi Gaussian")
plt.plot(t, _t, color="blue", label="$_t$")

plt.grid(True)
plt.legend()

plt.show()
```

### A.3.1 Urutan Unit Impuls

**Urutan Unit Impuls** adalah sebuah sinyal waktu diskrit urutan yang memiliki nilai (seperti bernilai 1) pada saat urutan tertentu. Urutan tersebut memiliki pola tidak seperti sinyal kontinu, namun lebih ke pola seperti bilangan bulat, bilangan ganjil, dan sejenisnya. 

Secara umum, urutan unit impuls memiliki bentuk seperti berikut ini:<br>
$[n] = 0; n$ lainnya<br>
$[n] = A; n = n \geq 0$

dengan $A$ adalah amplitudo impuls.

```python
# Pembuatan grafik

# deklarasi nilai setiap variabel
A = 2               # Amplitudo

# Setting plot untuk grafik
n = np.arange(-5, 5)
_n = np.where(n == 0, A, 0) 

# Plot grafik
plt.stem(n, _n, basefmt="None", label="$_n$")
plt.title("Grafik Urutan Unit Step")
plt.ylabel("amplitudo")
plt.xlabel("urutan (n)")
plt.ylim(-1, 3)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

$U(n) = 0; t < 0$<br>
$U(n) = A; t \geq 0$

dengan $A=$ tinggi nilai step.

```python
# Pembuatan grafik
# variabel
A = 3                   # Amplitudo
n = np.arange(-3, 6, 1) 

# Plot fungsi
U_n = np.where(n >= 0, A, 0)

plt.stem(n, U_n, label="$U(n)$", basefmt="None")
plt.title("Grafik Urutan Unit Step $U(n)$")
plt.ylabel("amplitudo")
plt.xlabel("urutan (n)")
plt.ylim(-1, 5)
plt.xlim(-4, 6)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
