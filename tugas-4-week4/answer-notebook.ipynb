## 1. Buktikan bahwa sinyal $x(t) = \sin(t)$ adalah sinyal periodik! 

Agar sinyal dikatakan periodik, kita harus membuktikan bahwa terdapat periode T sehingga $x(t) = \sin(t + T)$ masuk akal.

```python
# import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pembuatan grafik

# shortcut
sin = np.sin
pi = np.pi

# deklarasi variabel
t = np.linspace(0, 4 * pi, 1000)  # Meliputi dua periode, 0 - 4, pada fungsi x(t) = sin(t)

# plot fungsi
x_t = sin(t)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(t, x_t, label='x(t) = sin(t)')
plt.axvline(x=2*np.pi, color='r', linestyle='--', label='Periode = 2')
plt.axvline(x=4*np.pi, color='g', linestyle='--', label='Periode = 4')
plt.title('Sinyal x(t) = sin(t) dengan Periodicity')
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend()
plt.show()
```

Terlihat bahwa pada periode T tertentu, amplitudo sinyal masih tetap sama. 

## 2. Plot sinyal $x(t)= \cos(t) + \sin(3t)$ dalam 3 periode!

Sebelum itu, kita harus menentukan periode dari fungsi tersebut

fungsi $\cos(t)$
$T = \frac{2\pi}{\omega}$
$= \frac{2\pi}{1}$
$= 2\pi$

fungsi $\sin(3t)$
$T = \frac{2\pi}{\omega}$
$= \frac{2\pi}{3}$
$= \frac{2\pi}{3}$

KPK dari $2\pi$ dan $\frac{2\pi}{3}$ adalah $2\pi$ sehingga<br>
$T =2\pi$

```python
# Pembuatan grafik

# shortcut
cos = np.cos

# Deklarasi variabel
T = 2 * pi
j_periode = 3
t = np.linspace(0, j_periode * T, 1000)

# Plot sinyal 
x_t = cos(t) + sin(3 * t)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(t, x_t, label='x(t) = cos(t) + sin(3t)')
plt.axvline(x=T, color='r', linestyle='--', label='Period ke-1')
plt.axvline(x=2*T, color='g', linestyle='--', label='Period ke-2')
plt.axvline(x=3*T, color='b', linestyle='--', label='Period ke-3')
plt.title('Sinyal x(t) = cos(t) + sin(3t) dalam 3 periode')
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend()
plt.show()

```

## 3. Plot sinyal dan urutan dari fungsi kausal, non-kausal, dan anti-kausal


- Sinyal kausal adalah sinyal yang memiliki nilai $0$ saat $t < 0$
- Sinyal non-kausal adalah sinyal yang memiliki nilai di kedua keadaan $t < 0$ dan $t > 0$
- Sinyal anti-kausal adalah sinyal yang memiliki nilai $0$ saat $t > 0$

Mari kita buat grafik dengan fungsi $x(t) = e^{-t}$

```python
# Pembuatan grafik

# Deklarasi variabel
t = np.linspace(-10, 10, 1000)
t_disc = np.arange(-10, 10, 1)

# Original signal
x_t = sin(t) + cos(t)
x_t_disc = sin(t_disc) + cos(t_disc)

# Sinyal kausal: 0 saat t < 0
x_causal = np.where(t >= 0, x_t+1, 0)
x_causal_discrete = np.where(t_disc >= 0, x_t_disc+1, 0)

# sinyal non-kausal: terdapat nilai saat t < 0 and t > 0
x_noncausal = 2*x_t
x_noncausal_discrete = 2*x_t_disc

# sinyal Anti-kausal: 0 saat t > 0
x_anticausal = np.where(t <= 0, x_t-1, 0)
x_anticausal_discrete = np.where(t_disc <= 0, x_t_disc-1, 0)

# Plotting
plt.figure(figsize=(10, 8))

# Sinyal kontinu
plt.subplot(2, 1, 1)
plt.xlim(-12.5, 12.5)
plt.plot(t, x_causal, label='Sinyal kausal', color="red")
plt.plot(t, x_noncausal, label='Sinyal non-kausal', color="blue")
plt.plot(t, x_anticausal, label='Sinyal anti-kausal', color="green")
plt.title('Sinyal konitnu')
plt.grid(True)
plt.legend()

# Sinyal sequence
plt.subplot(2, 1, 2)
plt.xlim(-12.5, 12.5)
plt.stem(t_disc, x_causal_discrete, label='Sinyal kausal', linefmt="red", basefmt="none")
plt.stem(t_disc, x_noncausal_discrete, label='Sinyal non-kausal', linefmt="blue", basefmt="none",)
plt.stem(t_disc, x_anticausal_discrete, label='Sinyal anti-kausal', linefmt="green", basefmt="none")
plt.title('Sinyal sequence')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## 4. Buktikan bahwa sinyal $z[n] = \sin(\pi n + 2) + \cos(\frac{2\pi n}{3} +1)$ adalah sinyal periodik! 

Agar sinyal dikatakan periodik, kita harus membuktikan bahwa terdapat periode T sehingga $x(t) = \sin(t + T)$ masuk akal.

Periode $T = 6$

```python
# Pembuatan grafik
# deklarasi variabel
T = 6
n = np.arange(0, 4 * T, 1)  # Meliputi empat periode

# plot fungsi
z_n = sin((pi * n) +2) + cos((2 * pi * n / 3) + 1)

# Plotting
plt.figure(figsize=(8, 4))
plt.stem(n, z_n, label='x(t) = sin(t + 2) + cos((2n/3) + 1)')
plt.axvline(x=T, color='r', linestyle='--', label='Periode ke-1')
plt.axvline(x=2 * T, color='g', linestyle='--', label='Periode Ke-2')
plt.axvline(x=3 * T, color='b', linestyle='--', label='Periode Ke-3')
plt.title('Sinyal x(t) = sin(t + 2) + cos((2n/3) + 1) dengan Periodicity')
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend(fontsize="xx-small")
plt.show()
```

Terlihat bahwa pada periode T tertentu, amplitudo sinyal masih tetap sama. 

## 5. Periksa apakah sinyal $z[n] = \sin(\pi n + 2) + \cos(\frac{2n}{3} +1)$ adalah sinyal periodik atau aperiodik! 

Agar sinyal dikatakan periodik, kita harus membuktikan bahwa terdapat periode T sehingga $x(t) = \sin(t + T)$ masuk akal.

Mari kita plot fungsi dengan $t$ tinggi lalu mengecek apakah terdapat pola amplitudo yang sama pada *sequence* tertentu atau tidak.

```python
# Pembuatan grafik
# deklarasi variabel
n = np.arange(-25, 25, 1)  # Meliputi empat periode

# plot fungsi
z_n = sin((pi * n) +2) + cos((2 * n / 3) + 1)

# Plotting
plt.figure(figsize=(8, 4))
plt.stem(n, z_n, label='z[n] = sin(n + 2) + cos((2n/3) + 1)')
plt.title('Sinyal z[t] = sin(n + 2) + cos((2n/3) + 1) dengan Periodicity')
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend(fontsize="xx-small")
plt.show()
```

Walaupun agak susah terlihat, namun pada periode T hingga 25 detik setelah dana sebelumnya, amplitudo sinyal tidak ada yang sama. Jadi, sinyal ini adalah **aperiodic**

## 6. Periksa apakah sinyal $x(t) = t^2$ dan $y(t) = t^3$  adalah sinyal ganjil atau genap! 

Agar sinyal dikatakan ganjil atau genap, salah satu dari kriteria di bawah harus terpenuhi:<br>
Sinyal genap: $x(t) = x(-t)$; Umumnya simetri terhadap sumbu y tertentu.<br>
Sinyal ganjil: $x(t) = -x(-t)$; Umumnya simetri terbalik terhadap sumbu y.<br>

```python
# Pembuatan grafik
# deklarasi variabel
t = np.linspace(-20, 20, 1000)

# Sinyal x(t) = t^2
x_t2 = t**2

# Sinyal x(t) = t^3
x_t3 = t**3

# Plotting
plt.figure(figsize=(10, 6))

# Plot Even signal
plt.subplot(2, 1, 1)
plt.plot(t, x_t2, label='x(t) = t^2', color="red")
plt.title('x(t) = t^2')
plt.ylim(-100, 100)
plt.grid(True)
plt.legend()

# Plot Odd signal
plt.subplot(2, 1, 2)
plt.plot(t, x_t3, label='y(t) = t^3', color="blue")
plt.title('y(t) = t^3')
plt.ylim(-100, 100)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

Terlihat bahwa fungsi $x(t)$ simetri terhadap sumbu y sehingga fungsi tersebut adalah sinyal genap. 

Lalu, fungsi $y(t)$ simetri dengan kondisi terbalik terhadap sumbu y sehingga fungsi tersebut adalah sinyal genap.

## 7. Decompose sebuah sinyal step diskrit ke bentuk ganjil dan genap dengan interval $-17 < t < 17$ dan periksa kebenarannya dengan membangkitkan sinyal kontinyu nya! 

Diberikan fungsi step $u(t)$ yang didefinisikan dengan:<br>
$u(t) = 1$ untuk $t \geq 0$, dan<br>
$u(t) = 0$ untuk $t < 0$

Lalu, untuk mendekomposisi fungsi $u(t)$, gunakan rumus:<br>
$u_{genap}(t)=\frac{u(t)+u(-t)}{2}$<br>
$u_{ganjil}(t)=\frac{u(t)-u(-t)}{2}$<br>

```python
# Pembuatan grafik
# deklarasi variabel
n = np.arange(-17, 17, 1)

# deklarasi fungsi original
u_n = np.where(n >= 0, 1, 0)
u_t_negatif = np.where(n < 0, 1, 0)

# Fungsi Genap
u_genap = (u_n + u_t_negatif) / 2

# Fungsi Ganjil
u_ganjil = (u_n - u_t_negatif) / 2

# Plotting
plt.figure(figsize=(10, 6))

# Plot original unit step signal
plt.subplot(3, 1, 1)
plt.stem(n, u_n, label='u(n)', linefmt='r')
plt.title('Sinyal step diskrit u(n)')
plt.grid(True)
plt.ylim(-1.5, 1.5)
plt.legend()

# Plot Even part
plt.subplot(3, 1, 2)
plt.stem(n, u_genap, label='u(n) genap', linefmt='g')
plt.title('Bagian genap u(n)')
plt.grid(True)
plt.ylim(-1.5, 1.5)
plt.legend()

# Plot Odd part
plt.subplot(3, 1, 3)
plt.stem(n, u_ganjil, label='u(n) ganjil', linefmt='b')
plt.title('Bagian ganjil (n)')
plt.grid(True)
plt.ylim(-1.5, 1.5)
plt.legend()

plt.tight_layout()
plt.show()
```

Di bawah adalah bentuk kontinyu nya.

```python
# Pembuatan grafik
# deklarasi variabel
n = np.linspace(-17, 17, 500)

# deklarasi fungsi original
u_n = np.where(n >= 0, 1, 0)
u_t_negatif = np.where(n < 0, 1, 0)

# Fungsi Genap
u_genap = (u_n + u_t_negatif) / 2

# Fungsi Ganjil
u_ganjil = (u_n - u_t_negatif) / 2

# Plotting
plt.figure(figsize=(10, 6))

# Plot original unit step signal
plt.subplot(3, 1, 1)
plt.plot(n, u_n, label='u(n) (Unit Step)', color='r')
plt.title('Unit Step Signal u(n)')
plt.grid(True)
plt.ylim(-1.5, 1.5)
plt.legend()

# Plot Even part
plt.subplot(3, 1, 2)
plt.plot(n, u_genap, label='Even part of u(n)', color='g')
plt.title('Even Part of Unit Step Signal')
plt.grid(True)
plt.ylim(-1.5, 1.5)
plt.legend()

# Plot Odd part
plt.subplot(3, 1, 3)
plt.plot(n, u_ganjil, label='Odd part of u(n)', color='b')
plt.title('Odd Part of Unit Step Signal')
plt.grid(True)
plt.ylim(-1.5, 1.5)
plt.legend()

plt.tight_layout()
plt.show()
```

## 8. Periksa bahwa sinyal $x(t) = u(t)$ adalah power-type signal, sedangkan $x(t) = u(t)-u(t-1)$ adalah energy-type signal! 

```python
# Pembuatan grafik
# Deklarasi variable
t = np.linspace(-2, 5, 1000)

# Unit step function u(t)
u_t = np.where(t >= 0, 1, 0)

# Pulse function u(t) - u(t-1)
pulse = np.where((t >= 0) & (t < 1), 1, 0)

# Compute energy of the pulse (u(t) - u(t-1))
energy_pulse = np.trapz(np.abs(pulse)**2, t)

# Compute power of the unit step u(t)
# Since it's infinite, we approximate power over a finite range
power_step = np.mean(np.abs(u_t)**2)

# Plotting
plt.figure(figsize=(10, 6))

# Plot unit step function u(t)
plt.subplot(2, 1, 1)
plt.plot(t, u_t, label='u(t) (Unit Step)')
plt.title(f'Unit Step Signal u(t) (Approx. Power = {power_step:.2f})')
plt.grid(True)
plt.legend()

# Plot pulse function u(t) - u(t-1)
plt.subplot(2, 1, 2)
plt.plot(t, pulse, label='u(t) - u(t-1) (Pulse)')
plt.title(f'Pulse Signal u(t) - u(t-1) (Energy = {energy_pulse:.2f})')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## 9. Kalkulasi energy dan power dari sinyal $x(t) =2\cos(\pi t)$ dalam satu periode!

Fungsi $x(t) =2\cos(\pi t)$ memiliki periode:<br>
$T=2$

```python
# Pembuatan grafik
# Deklarasi variable
T = 2
t = np.linspace(0, T, 1000)

# Deklarasi fungsi
x_t = 2 * cos(pi * t)

# Kalkulasi energi
energi_x_t = np.trapz(np.abs(x_t)**2, t)

# Kalkulasi power
# Since it's infinite, we approximate power over a finite range
power_x_t = np.mean(np.abs(x_t)**2)

# Plotting
plt.figure(figsize=(10, 6))

# Plot unit step function u(t)
plt.subplot(2, 1, 1)
plt.plot(t, x_t, label='x(t)')
plt.title(f'x(t) = 2cos(t)(Approx. Power = {power_x_t:.2f}; Energy = {energi_x_t:.2f})')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## 12. Plot sinyal $x(-t)$ dengan interval $-1 \leq t \leq 3$ dan $x(t)=te^{-t}$

```python
# Pembuatan grafik
# Deklarasi variable
t = np.linspace(-1, 3, 1000)

# Sinyal original
def u_t(t) -> list:
    x_t = t * exp(-t)
    return x_t

# Sinyal yang akan dibangkitkan
x_t = u_t(t)
x_t_negate = u_t(-t)

# Kalkulasi energi

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(t, x_t, label="x(t)")
plt.plot(t, x_t_negate, label="x(-t)")
plt.title(f'x(-t) dari x(t) = te^(-t)')
plt.xlabel('waktu(t)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.legend()
plt.show()
```

## 13. Menggunakan fungsi dan interval yang sama pada no 12, plot sebuah sinyal $x_1(t)=x(2t)$ dan $x_2(t)=x(\frac{1}{2}t)$

```python
# Pembuatan grafik
# Deklarasi variable
t = np.linspace(-1, 3, 1000)

# Sinyal original
def u_t(t) -> list:
    x_t = t * exp(-t)
    return x_t

# Sinyal yang akan dibangkitkan
x_1 = u_t(2 * t)
x_2 = u_t(0.5 * t)

# Kalkulasi energi

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(t, x_1, label="x_1(t) = x(2t)", color="r")
plt.title("x_1(t)")
plt.xlabel('waktu(t)')
plt.ylabel('Amplitudo')
plt.ylim(-15, 2)
plt.grid(True)
plt.legend()

plt.figure(figsize=(8, 4))
plt.plot(t, x_2, label="x_2(t)= x(0.5t)", color="g")
plt.title('x_2(t)')
plt.xlabel('waktu(t)')
plt.ylabel('Amplitudo')
plt.ylim(-15, 2)
plt.grid(True)
plt.legend()

plt.show()
```
