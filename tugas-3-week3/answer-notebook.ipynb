## 1. Plot sebuah sinyal waktu kontinu $x(t) = \cos(t)$ dalam interval $-n \leq t \leq n$ 

```python
# import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
import string
```

```python
# Pembuatan grafik
# Shortcut
cos = np.cos

# Deklarasi nilai tiap variabel
n = 17
t = np.linspace(-n, n, 1000)

# Setting plot untuk grafik
x_t = cos(t)

# Display grafik
plt.figure(figsize=(10, 5))
plt.ylim(-2, 2)
plt.xlim(-18, 18)
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.plot(t, x_t, color="green")
plt.grid(True)
```

## 4. Plot sebuah sinyal digital $x[n] = \cos(n)$ dalam interval $-n \leq t \leq n$ 

Asumsikan sinyal dibuat dalam 1-bit. 

Nilai yang mungkin dari grafik hanya 0 dan 1.<br>
Untuk $\cos(n) < 0$, $x(n) = 0$, dan <br>
Untuk $\cos(n) \geq 0$, $x(n) = 1$.

```python
# Pembuatan grafik
# Deklarasi nilai tiap variabel
n = 17
t_digit = np.arange(-n, n, 1)

# Setting plot untuk grafik
x_t_disc = cos(t_digit)
x_t_digit = np.where(x_t_disc >= 0, 1, 0)

# Display grafik
plt.figure(figsize=(10, 5))
plt.ylim(-2, 2)
plt.xlim(-18, 18)
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")
plt.stem(t_digit, x_t_digit, linefmt="blue", basefmt="none", markerfmt="blue")
plt.grid(True)
```

## 7. Plot sebuah sinyal waktu kontinu $x(t) = Ae^{\beta _1t}$ dan $y(t) = Ae^{\beta _2t}$ dalam satu grafik dengan interval $-T \leq t \leq T$, di mana $A$ ditentukan, $\beta _1 = \frac{A}{100}$, $\beta _2 = -\frac{A}{100}$, dan $T = \frac{A}{2}$   

```python
# Pembuatan grafik
# Shortcut
exp = np.exp

# Deklarasi nilai tiap variabel
A = 17
beta_1 = A/100
beta_2 = -(A/100)
T = A/2
t = np.linspace(-T, T, 1000)

# Setting plot untuk grafik
x_t = A * exp(beta_1 * t)
y_t = A * exp(beta_2 * t)

# Konfigurasi sub plot
fig, axs = plt.subplots(3, 1, figsize=(10,12))

# Grafik x(t)
axs[0].set_title("Grafik Fungsi x(t)")
axs[0].set_xlabel("waktu (s)")
axs[0].set_ylabel("Amplitudo")
axs[0].plot(t, x_t, color="red")
axs[0].grid(True)

# Grafik y(t)
axs[1].set_title("Grafik Fungsi y(t)")
axs[1].set_xlabel("waktu (s)")
axs[1].set_ylabel("Amplitudo")
axs[1].plot(t, y_t, color="blue")
axs[1].grid(True)

# Grafik gabungan
axs[2].set_title("Grafik gabungan")
axs[2].set_xlabel("waktu (s)")
axs[2].set_ylabel("Amplitudo")
axs[2].plot(t, y_t, color="blue", label="x(t)")
axs[2].plot(t, x_t, color="red", label="y(t)")
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

## 8. Plot sinyal real dan imajiner dari fungsi $y(t)=2e^{j\pi t +\frac{{\pi}}{3}}$ dalam satu periode

fungsi $y(t)=2e^{j\pi t +\frac{{\pi}}{3}}$ jika diubah dalam bentuk lain dapat ditulis sebagai berikut:

$y(t)=2e^{j\pi t +\frac{{\pi}}{3}}$<br>
$= 2[\cos(\pi t+\frac{{\pi}}{3}) + j\sin(\pi t+\frac{{\pi}}{3})] $<br>

dengan<br>
$y(t)_{real}        = 2\cos(\pi t+\frac{{\pi}}{3}) $<br>
$y(t)_{imajiner}    = 2\sin (\pi t+\frac{{\pi}}{3}) $<br>

Periode dapat dicari dengan menelaah fungsi dan mencari nilai $\omega = \frac{2\pi}{T}$, di mana $\omega = \pi$ sehingga

$T = \frac{2\pi}{\omega} = \frac{2\pi}{\pi} = 2$ 

```python
# Pembuatan grafik
# Deklarasi nilai tiap variabel
T = 2
t = np.linspace(0, T, 1000)

# Setting plot untuk grafik
y_real      = 2 * cos (pi * t + (pi/3))
y_imajiner  = 2 * sin (pi * t + (pi/3))

# Display
plt.figure(figsize=(10, 5))
plt.ylim(-3, 3)
plt.xlim(-3, 3)
plt.xlabel("waktu(s)")
plt.ylabel("amplitudo")

# Grafik y_real
plt.plot(t, y_real, color="green", label="y(t)_real")
# Grafik y_imajiner
plt.plot(t, y_imajiner, color="blue", label="y(t)_imajiner")

plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of n (time or sample index)
n = np.linspace(0, 4 * 2 * np.pi, 1000)  # 1000 points covering 4 periods

# Generate the cosine signal with four periods
x_n = np.cos(n)

# Plotting the cosine signal
plt.plot(n, x_n, label='Cosine Signal')

# Add vertical lines at the end of each period (every 2π)
for i in range(1, 5):
    plt.axvline(x=i * 2 * np.pi, color='red', linestyle='--', label=f'End of Period {i}')

# Labeling and formatting
plt.title("Cosine Signal with Four Periods")
plt.xlabel("n (samples)")
plt.ylabel("x[n] = cos(n)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define time array for one period
T = 2  # Period of the signal
t = np.linspace(0, T, 1000)  # 1000 points from 0 to T

# Compute the real and imaginary parts of the signal
y_real = 2 * np.cos(np.pi * t + np.pi/3)
y_imag = 2 * np.sin(np.pi * t + np.pi/3)

# Plot the real part
plt.figure(figsize=(10, 6))
plt.plot(t, y_real, label='Real part: 2cos(πt + π/3)', color='b')

# Plot the imaginary part
plt.plot(t, y_imag, label='Imaginary part: 2sin(πt + π/3)', color='r')

# Add labels and title
plt.xlabel('Time t (seconds)')
plt.ylabel('Amplitude')
plt.title('Real and Imaginary Parts of the Signal y(t) = 2e^{j(πt + π/3)}')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
