## 1. Sebuah sistem MIMO dideskripsikan sebagai hubungan input output dari $y_1(t) = x_1(t) + x_2(t)$ dan $y_2 = x_1(t) - x_2(t).$ <br><br> Kalkulasikan dan plotkan respon dari sistem tersebut untuk $x_1(t)=u(t)$ dan $x_2(t)=0.5u(t)$!

```python
# import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pembuatan grafik
# deklarasi variabel
t = np.linspace(-5, 10, 1000)
def u(t):
    return np.where(t > 0, 1, 0)

# plot fungsi
x_t1 = u(t)
x_t2 = 0.5 * u(t)

y_t1 = x_t1 + x_t2
y_t2 = x_t1 - x_t2

# Plotting 
plt.figure(figsize=(8,4))

# grafik y_1t
plt.subplot(2, 1, 1)
plt.xlim(-5, 11)
plt.ylim(0, 2)
plt.title("Respon grafik $y_1(t)$")
plt.grid(True)
plt.plot(t, y_t1, color='r')

# grafik y_2t
plt.subplot(2, 1, 2)
plt.xlim(-5, 11)
plt.ylim(0, 2)
plt.title("Respon grafik $y_2(t)$")
plt.grid(True)
plt.plot(t, y_t2, color='b')

plt.tight_layout()
plt.show()
```

## 2. Sebuah sistem MIMO dideskripsikan sebagai hubungan input output dari $y(t) = x_1(t) + x_2(t).x_3(t)$ <br><br> Kalkulasikan dan plotkan respon dari sistem tersebut untuk $x_1(t)=u(t) - u(t - 3)$, $x_2(t)=t\sin(t), 0 \leq t \geq 4$, dan $x_3(t) = t\cos(t), 0 \leq t \geq 4$!

```python
# Pembuatan grafik

# shortcut
sin = np.sin
cos = np.cos

# deklarasi variabel
t = np.linspace(-5, 10, 1000)
def u(t):
    return np.where(t > 0, 1, 0)

# plot fungsi
x_t1 = u(t) - u(t-3)
x_t2 = np.where((t <= 0) | (t >=4), t* sin(t), 0)
x_t3 = np.where((t <= 0) | (t >=4), t* cos(t), 0)

y_t = x_t1 + x_t2 * x_t3
# Plotting 
plt.figure(figsize=(10,15))

# grafik x1(t)
plt.subplot(4, 1, 1)
plt.xlim(-5, 11)
plt.ylim(-10, 10)
plt.title("Input Grafik $x_1(t) = u(t) - u(t-3)$")
# plt.axvline(x=3, color='', linestyle="--", label="t = 3")
plt.axvline(x=0, color='#565656', linestyle="-")
plt.grid(True)
plt.plot(t, x_t1, color='b')

# grafik x2(t)
plt.subplot(4, 1, 2)
plt.xlim(-5, 11)
plt.ylim(-10, 10)
plt.title("Input Grafik $x_2(t) = t.\\sin(t)$ \n $0 \leq t \geq 4$")
# plt.axvline(x=3, color='y', linestyle="--", label="t = 3")
plt.axvline(x=0, color='#565656', linestyle="-")
plt.grid(True)
plt.plot(t, x_t2, color='y')

# grafik x3(t)
plt.subplot(4, 1, 3)
plt.xlim(-5, 11)
plt.ylim(-10, 10)
plt.title("Input grafik $x_3(t) = t.\cos(t)$ \n $0 \leq t \geq 4$")
# plt.axvline(x=3, color='y', linestyle="--", label="t = 3")
plt.axvline(x=0, color='#565656', linestyle="-")
plt.grid(True)
plt.plot(t, x_t3, color='g')

# grafik yt
plt.subplot(4, 1, 4)
plt.xlim(-5, 11)
plt.ylim(-10, 10)
plt.title("Respon grafik $y(t)$")
plt.axvline(x=3, color='y', linestyle="--", label="t = 3")
plt.axvline(x=4, color='g', linestyle="--", label="t = 4")
plt.axvline(x=0, color='#565656', linestyle="-")
plt.grid(True)
plt.plot(t, y_t, color='r')

plt.tight_layout()
plt.legend()
plt.show()
```

## 3. Sebuah sistem MIMO dideskripsikan sebagai hubungan input output dari $y_1(t) = x_1(t) + x_2(t)$ dan $y_2 = x_1(t) - x_2(t).$ <br><br> Kalkulasikan dan plotkan respon dari sistem tersebut untuk $x_1(t)=u(t)$ dan $x_2(t)=0.5t\times u(t-1)$!

```python
# Pembuatan grafik
# deklarasi variabel
t = np.linspace(-5, 10, 1000)
def u(t):
    return np.where(t > 0, 1, 0)

# plot fungsi
x_t1 = u(t)
x_t2 = 0.5 * t * u(t-1)

y_t1 = x_t1 + x_t2
y_t2 = x_t1 - x_t2

# Plotting 
plt.figure(figsize=(8,4))

# grafik y_1t
plt.subplot(2, 1, 1)
plt.xlim(-5, 11)
plt.ylim(-5, 5)
plt.title("Respon grafik $y_1(t)$")
plt.grid(True)
plt.plot(t, y_t1, color='r')

# grafik y_2t
plt.subplot(2, 1, 2)
plt.xlim(-5, 11)
plt.ylim(-5, 5)
plt.title("Respon grafik $y_2(t)$")
plt.grid(True)
plt.plot(t, y_t2, color='b')

plt.tight_layout()
plt.show()
```

## 4. Sebuah sistem MIMO dideskripsikan sebagai hubungan input output dari $y(t) = x_1(t) +[ x_2(t) + x_3(t)]$ <br><br> Kalkulasikan dan plotkan respon dari sistem tersebut untuk $x_1(t)=e^{2t}, 0 \leq t \leq 5$, $x_2(t)=te^{t}, 0 \leq t \leq 5$, dan $x_3(t) = \cos(t), 0 \leq t \leq 5$!

```python
# Pembuatan grafik
# shortcut
exp = np.exp

# deklarasi variabel
t = np.linspace(-5, 10, 1000)

# plot fungsi
x_t1 = np.where((0 <= t) & (t <= 5), exp(2 * t), 0)
x_t2 = np.where((0 <= t) & (t <= 5), t * exp(t), 0)
x_t3 = np.where((0 <= t) & (t <= 5), cos(t), 0)

y_t = x_t1 * (x_t2 + x_t3)
# Plotting 
plt.figure(figsize=(10,15))

# grafik x1(t)
plt.subplot(4, 1, 1)
plt.xlim(-5, 11)
plt.ylim(-3, 25)
plt.title("Input Grafik $x_1(t) = e^{2t}$, 0 <= t <= 5")
# plt.axvline(x=3, color='', linestyle="--", label="t = 3")
plt.axvline(x=0, color='#565656', linestyle="-")
plt.grid(True)
plt.plot(t, x_t1, color='b')

# grafik x2(t)
plt.subplot(4, 1, 2)
plt.xlim(-5, 11)
plt.ylim(-3, 25)
plt.title("Input Grafik $x_2(t) = t.\\sin(t)$ \n $0 \leq t \geq 4$")
# plt.axvline(x=3, color='y', linestyle="--", label="t = 3")
plt.axvline(x=0, color='#565656', linestyle="-")
plt.grid(True)
plt.plot(t, x_t2, color='y')

# grafik x3(t)
plt.subplot(4, 1, 3)
plt.xlim(-5, 11)
plt.ylim(-3, 25)
plt.title("Input grafik $x_3(t) = t.\cos(t)$ \n $0 \leq t \geq 4$")
# plt.axvline(x=3, color='y', linestyle="--", label="t = 3")
plt.axvline(x=0, color='#565656', linestyle="-")
plt.grid(True)
plt.plot(t, x_t3, color='g')

# grafik yt
plt.subplot(4, 1, 4)
plt.xlim(-5, 11)
plt.ylim(-3, 25)
plt.title("Respon grafik $y(t)$")
# plt.axvline(x=3, color='y', linestyle="--", label="t = 3")
# plt.axvline(x=4, color='g', linestyle="--", label="t = 4")
plt.axvline(x=0, color='#565656', linestyle="-")
plt.grid(True)
plt.plot(t, y_t, color='r')

plt.tight_layout()
plt.legend()
plt.show()
```

## 5. Sebuah sistem MIMO dideskripsikan sebagai hubungan input output dari $y_1[n] = x_1[n] / x_2[n]$ dan $y_2[n] = x_1[n] \times x_2[n].$ <br><br> Kalkulasikan dan plotkan respon dari sistem tersebut untuk $x_1[n]=u[n]$ dan $x_2[n]=n^2$!

```python
# Pembuatan grafik
# deklarasi variabel
n = np.arange(-12, 12, 1)
def u(n):
    return np.where(n > 0, 1, 0)

# plot fungsi
x_n1 = np.where((-10 <= n) & (n <= 10), u(n), 0)
x_n2 = np.where((-10 <= n) & (n <= 10), n^2, 0)
                 
y_n1 = x_n1 / x_n2
y_n2 = x_n1 * x_n2

# Plotting 
plt.figure(figsize=(10,15))

# grafik y_1t
plt.subplot(3, 1, 1)
plt.xlim(-12, 12)
plt.title("Respon grafik $y_1[n]$")
plt.grid(True)
plt.stem(n, y_n1, linefmt="r")

# grafik y_2t
plt.subplot(3, 1, 2)
plt.xlim(-12, 12)
plt.title("Respon grafik $y_2[n]$")
plt.grid(True)
plt.stem(n, y_n2, linefmt="b")

# grafik gabungan
plt.subplot(3, 1, 3)
plt.xlim(-12, 12)
plt.ylim(-2, 12)
plt.title("Respon grafik gabungan")
plt.grid(True)
plt.stem(n, y_n2, linefmt="b")
plt.stem(n, y_n1, linefmt="r")

plt.tight_layout()
plt.show()
```
